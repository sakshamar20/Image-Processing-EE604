```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import cv2
from io import BytesIO
np.set_printoptions(threshold=np.inf)
```

```python
def show(image):
    cv2.imshow("Image", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def convert():
    buffer = BytesIO()
    plt.savefig(buffer, format="png")
    buffer.seek(0)

# Read the image from the buffer using OpenCV
    image = cv2.imdecode(np.frombuffer(buffer.read(), np.uint8), -1)
    return image
```

```python
# path = 'D:\Assignment 1\Q2\\test'
# audio_path= [filename for filename in os.listdir(path) if filename.endswith(".mp3")]

audio_path = "D:\Assignment 1\Q2\\test\metal_banging1.mp3"
y, sr = librosa.load(audio_path, sr=None)
n_fft = 2048
hop_length = 512
spec = librosa.feature.melspectrogram(y=y,sr=sr,n_fft=n_fft, hop_length = hop_length, fmax = 22000)

spec_db= np.abs(spec)**2
spec_db = db_spectrogram = librosa.power_to_db(spec_db, ref=np.max)
plt.figure()
librosa.display.specshow(spec_db, y_axis='log', x_axis='time')
```

```python
fspec = spec.ravel()
np.shape(fspec)
x = np.linspace(1,33280,33280)
print(np.shape(fspec), np.shape(x))
plt.plot(x,fspec)
```

```python
fspec = spec.ravel()
np.shape(fspec)
x = np.linspace(1,33280,33280)
print(np.shape(fspec), np.shape(x))
plt.plot(x,fspec)
np.mean(fspec)
```
