```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
show(gray)
print(np.shape(gray))
```

```python
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

# Calculate the absolute gradients
abs_sobel_x = np.absolute(sobel_x)
abs_sobel_y = np.absolute(sobel_y)

# Combine the gradient images
gradient = cv2.addWeighted(abs_sobel_x, 0.5, abs_sobel_y, 0.5, 0)

# Threshold the gradient to create a binary edge map
threshold_value = 50  # Adjust this threshold as needed
edge_map = cv2.threshold(gradient, threshold_value, 255, cv2.THRESH_BINARY)[1]
show(edge_map)
```

```python
h, theta, d = cv2.HoughLines(edge_map)
accum, angles, dists = cv2.HoughLinesPointSet(h, theta, d)
angle = np.rad2deg(np.mode(angles)[0][0])

if (angle < 0):

    angle = angle + 90

else:

    angle = angle - 90

print(angle)
```
